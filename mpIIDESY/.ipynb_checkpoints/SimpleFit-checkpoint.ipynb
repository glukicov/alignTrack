### Function $N(t)=Ne^{-t/\tau}[1+A\cos(\omega_at+\phi)]$,
where  
[0] $N$ is overall normalisation  
[1] $\tau$ is the boosted muon lifetime $\tau = \gamma \cdot \tau_0 = 29.3\cdot2.2=66.44 \, \mu$s  
[2] $A$ is the asymmetry  
[3] $\omega_a$ is the anomalous precision frequency  
[4] $\phi$ is the initial phase  

```python
legend =  TLegend(0.3,0.55,0.90,0.88)
f = TF1("f", "[0]*exp(-x*[1]) * (1+[2]*cos([3]*x+[4]) )", xmin, xmax, 5)
f.SetNpx(10000) # prevent under-sampling
f.SetParLimits(3, 1.4, 1.5) # w_a 
f.SetLineColor(3) # green! 
h.Fit(f, "R")
h.Draw()
f.Draw("same")
chi2 =f.GetChisquare()
ndf=f.GetNDF();
ch2ndf=chi2/ndf
# print("Chi2/ndf",chi2/ndf,"\n")
# print("Entries", entries)

LT = 1/f.GetParameter(1)
LT_error = f.GetParError(1)
phi = 1/f.GetParameter(4)
phi_error = f.GetParError(4)
# print(LT, LT_error)

legend.SetHeader(r"16145-16234\,(\mathrm{HighKick}) \ N(t)=Ne^{-t/\tau} \cdot [1+A\cos(\omega_{a}t+\phi)]", "C")
legenValue1 = "#chi^{2}_{ndf}=%.2f #tau=%.2f #mus #phi=%.3f #pm %.3f" % (ch2ndf, LT, psi, psi_error)
legenValue2 = "Entries (30 #mus - 150 #mus; > 1.8 GeV): %.2E" % Decimal(entries)
# legenValue1 = "#splitline{"+str(names[i_state])+"}{"+plotMean[i_plot]+": "+str(round(mean, round_to))+" #pm "+str(round(mean_error,round_to))+"}"
# legenValue2 = "#sigma: "+str(round(sd,round_to))+" #pm "+str(round(sd_error,round_to))
legend.AddEntry(f, "#splitline{"+str(legenValue1)+"}{"+str(legenValue2)+"}", "")
legend.Draw("same")
gStyle.SetLegendBorderSize(0)
gStyle.SetLegendTextSize(0.044)

h.GetXaxis().SetRangeUser(xmin-5, xmax)
c.Draw()
c.SaveAs("wiggle.png")
```
